```python
import pandas as pd
import pycountry
```

```python
gap=pd.read_csv('gapminder.csv')
gap
```

```python
pycountry.countries.get(name='Tokelau')
```

```python
def codigo(pais, code=2):
    try:
        if code==2:
            return pycountry.countries.get(name=pais).alpha_2
        else:
            return pycountry.countries.get(name=pais).alpha_3
    except:
        return None
```

```python
codigo('St.-Pierre-et-Miquelon', 3)
```

```python
gap['alpha_2'] = gap['country'].apply(lambda x: codigo(x))
gap['alpha_3'] = gap['country'].apply(lambda x: codigo(x, 3))
```

```python
gap
```

```python
gap.dropna(inplace=True)
```

```python
gap.to_csv('gap.csv', index=False)
```
